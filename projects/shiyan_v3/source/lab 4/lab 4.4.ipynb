```python
from math import sqrt
from numpy import dot
from numpy.linalg import norm

def cos_sim(a, b): 
    return dot(a, b) / (norm(a) * norm(b))
```

```python
from gensim.models import Word2Vec

model = Word2Vec.load("data\\w2v_model")
```

```python
words = list(model.wv.index_to_key)

print(words)
```
